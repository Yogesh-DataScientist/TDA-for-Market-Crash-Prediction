```python
import yfinance as yf
import pandas as pd
import os

os.makedirs("data", exist_ok=True)

tickers = {
    "SP500": "^GSPC",
    "VIX": "^VIX",
    "NASDAQ": "^IXIC",
    "GOLD": "GC=F",
    "OIL": "CL=F"
}

all_data = {}

for name, ticker in tickers.items():
    print(f"Downloading {name}...")

    df = yf.download(
        ticker,
        start="2000-01-01",
        auto_adjust=False,
        progress=False
    )

    print(f"{name}: {df.shape}")

    if not df.empty:
        all_data[name] = df["Adj Close"].squeeze()

market_data = pd.DataFrame(all_data)

market_data.index.name = "Date"

print("\nDataset Shape:")
print(market_data.shape)

print("\nMissing Values:")
print(market_data.isnull().sum())

print("\nFirst 5 rows:")
print(market_data.head())

print("\nLast 5 rows:")
print(market_data.tail())

market_data.to_csv("data/market_data_raw.csv")

print("\nSaved: data/market_data_raw.csv")
```
